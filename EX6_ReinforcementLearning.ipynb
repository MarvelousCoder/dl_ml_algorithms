```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
dataset.head(10)
 
```

```python
import random
N = 10000 # num de exemplos
d = 10 # num de ads por exemplo
ads_selected = []
numbers_of_rewards_1 = [0] * d
numbers_of_rewards_0 = [0] * d
total_reward = 0
for n in range(0, N):   # vai para o prox exemplo
    # ARRISCA UM AD QUE PARECA TER CHANCE DE RECOMPENSA
    ad = 0
    max_random = 0
    for i in range(0, d):  # analisa os ads do exemplo atual
        # verifica qual tem mais chance de recompensa
        random_beta = random.betavariate(numbers_of_rewards_1[i] + 1, numbers_of_rewards_0[i] + 1) # numero sorteado
        if random_beta > max_random:
            max_random = random_beta
            ad = i  # anota o ad com mais chance de recompensa no exemplo atual
    ads_selected.append(ad)  # anota os ads tentados 
    # FEEDBACK
    reward = dataset.values[n, ad] # verifica se houve recompensa para o ad em que estimou mais chance de recompensa
    if reward == 1:
        numbers_of_rewards_1[ad] = numbers_of_rewards_1[ad] + 1   # anota historico de recompensas para o ad em que viu chance
    else:
        numbers_of_rewards_0[ad] = numbers_of_rewards_0[ad] + 1   # anota historico de nao recompensas para o ad em que viu chance
    total_reward = total_reward + reward  # anota o total de recompensas ja conseguido

#When we run and the code and visualize:	
plt.hist(ads_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
